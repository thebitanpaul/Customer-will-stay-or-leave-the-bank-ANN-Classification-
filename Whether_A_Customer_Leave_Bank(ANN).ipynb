##Data Preprocessing

```python
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
df
```

```python
x = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoding', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

#Building The ANN

###Initialize The ANN

```python
ann = tf.keras.models.Sequential()
```

###Adding Input Layer & First Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

we used 'relu' (rectified) activation function bcz in case of fully connected ann, we gotta use rectified activation function within the hidden layers

###Adding Second Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

###Adding Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

we used 'sigmoid' activation function bcz it gives us the output as well as the probablity of the output

#Traing The Ann

###Compiling The Ann

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

For binary classfication(yes or no 0/1), we use loss = 'binary_crossentropy' & activation='sigmoid' 

For multiclass classfication(east, west, north), we use loss = 'categorical_crossentropy' & activation='softmax' 


but in case of ANN regression we use loss = 'mean_squared_error' & no activation function

###Training ANN On The Training Set

```python
ann.fit(x_train, y_train, batch_size=32, epochs=100)
```

here, batch_size defines the set of predicted values to be compared simultaniously with the actual output values

epochs means one complete forward & backward propagation of the data through the ANN

##Pedicting A Given Perticular Test Value

**Homework**

Use our ANN model to predict if the customer with the following informations will leave the bank: 

Geography: France

Credit Score: 600

Gender: Male

Age: 40 years old

Tenure: 3 years

Balance: \$ 60000

Number of Products: 2

Does this customer have a credit card ? Yes

Is this customer an Active Member: Yes

Estimated Salary: \$ 50000

So, should we say goodbye to that customer ?

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])))
```

Therefore, our ANN model predicts that this customer stays in the bank!

**Important note 1:** Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array.

**Important note 2:** Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of features X, "France" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

**Important note 3: As the training data was scalled, we need to scale the test input also

**Important note 4: As we used 'sigmoid' activation method in our output layer, so the predicted values is in the form of probablity

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) >0.5)
```

To convert the probablity into yes/no classification, we used a logic >0.5
